## Seir Model
<b>S</b>usceptible <br>
<b>E</b>xposed <br>
<b>I</b>nfected<br>
<b>R</b>ecovered<br>



1.   α is the inverse of the incubation period (1/t_incubation)
2.   β is the average contact rate in the populatioN
3.γ is the inverse of the mean infectious period (1/t_infectious)


https://towardsdatascience.com/social-distancing-to-slow-the-coronavirus-768292f04296


```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import math
```

```
# Define parameters
t_max = 365
dt = 1
t = np.linspace(0, t_max, int(t_max/dt) + 1)
N = 12252023
init_vals = 1 - 1/N, 1/N, 0, 0
alpha = 0.2
beta = 2
gamma = 0.072
params = alpha, beta, gamma
```

```
def base_seir_model(init_vals, params, t):
   
    S_0, E_0, I_0, R_0 = init_vals
    S, E, I, R = [S_0], [E_0], [I_0], [R_0]
    alpha, beta, gamma = params
    listbeta = [beta]
    dt = t[1] - t[0]
    for _ in t[1:]:
        next_S = S[-1] - (beta*S[-1]*I[-1])*dt
        next_E = E[-1] + (beta*S[-1]*I[-1] - alpha*E[-1])*dt
        next_I = I[-1] + (alpha*E[-1] - gamma*I[-1])*dt
        next_R = R[-1] + (gamma*I[-1])*dt
        S.append(next_S)
        E.append(next_E)
        I.append(next_I)
        R.append(next_R)
        listbeta.append(beta)
        if next_I > 0.1:
          if beta <= 0.5:
            0.5
          else:
            beta = (beta/math.exp(_/1000))
        else: beta 
    return np.stack([S, E, I, R,listbeta]).T
```

```

# Run simulation
results = base_seir_model(init_vals, params, t)

#plt.plot(results)
#plt.show()
```

```
df =pd.DataFrame(results)
df.columns = ['S','E','I','R','beta']
df['Ipopulation'] = df['I']*N
df['Epopulation'] = df['E']*N
#df['S']=df.S.astype('int64')
#df['E']=df.E.astype('int64')
#df['I']=df.I.astype('int64')
#df['R']=df.R.astype('int64')

df.loc[27]
```
